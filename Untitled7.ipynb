```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms

# Provided Fisher matrix
fisher_matrix = np.array([
    [2.54444e+10, -3.55773e+09, 3.63459e+02, -2.68738e+00, -1.11757e+09, -3.86419e+06],
    [-3.55773e+09, 6.60399e+08, -2.29731e+02, 3.13757e+00, 3.97446e+08, 8.33542e+05],
    [3.63459e+02, -2.29731e+02, 3.41294e-03, -9.41965e-05, -1.73880e+03, -6.23024e-01],
    [-2.68738e+00, 3.13757e+00, -9.41965e-05, 2.74312e-06, 4.22477e+01, 1.10692e-02],
    [-1.11757e+09, 3.97446e+08, -1.73880e+03, 4.22477e+01, 1.18004e+09, 7.63258e+05],
    [-3.86419e+06, 8.33542e+05, -6.23024e-01, 1.10692e-02, 7.63258e+05, 1.15600e+03]
])

# Invert the Fisher matrix to get the covariance matrix
cov_matrix = np.linalg.inv(fisher_matrix)

# Function to plot confidence ellipse
def plot_confidence_ellipse(ax, mean, cov, n_std=1.0, facecolor='none', **kwargs):
    pearson = cov[0, 1] / np.sqrt(cov[0, 0] * cov[1, 1])
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0),
                      width=ell_radius_x * 2,
                      height=ell_radius_y * 2,
                      facecolor=facecolor,
                      **kwargs)
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = mean[0]
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = mean[1]
    transf = transforms.Affine2D().rotate_deg(45).scale(scale_x, scale_y).translate(mean_x, mean_y)
    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)

# Parameters to plot
param_indices = [(0, 1), (0, 2), (1, 2), (3, 4), (4, 5)]  # Example pairs (M_c, eta), (M_c, Lambda), etc.
param_names = ['M_c', 'eta', 'Lambda', 'delta Lambda', 't_c', 'phi_c']

# Plot confidence ellipses
fig, axes = plt.subplots(len(param_indices), figsize=(10, 15))

for i, (param1, param2) in enumerate(param_indices):
    ax = axes[i]
    mean = [0, 0]  # Assuming the parameters are centered at zero
    cov = cov_matrix[[param1, param2], :][:, [param1, param2]]

    plot_confidence_ellipse(ax, mean, cov, n_std=1, edgecolor='blue', label='1$\sigma$')
    plot_confidence_ellipse(ax, mean, cov, n_std=2, edgecolor='green', label='2$\sigma$')
    plot_confidence_ellipse(ax, mean, cov, n_std=3, edgecolor='red', label='3$\sigma$')

    ax.scatter(0, 0, c='red')  # Mean point
    ax.set_xlabel(param_names[param1])
    ax.set_ylabel(param_names[param2])
    ax.legend()
    ax.set_title(f'Confidence ellipses for {param_names[param1]} and {param_names[param2]}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import seaborn as sns

# Provided Fisher matrix
fisher_matrix = np.array([
    [2.54444e+10, -3.55773e+09, 3.63459e+02, -2.68738e+00, -1.11757e+09, -3.86419e+06],
    [-3.55773e+09, 6.60399e+08, -2.29731e+02, 3.13757e+00, 3.97446e+08, 8.33542e+05],
    [3.63459e+02, -2.29731e+02, 3.41294e-03, -9.41965e-05, -1.73880e+03, -6.23024e-01],
    [-2.68738e+00, 3.13757e+00, -9.41965e-05, 2.74312e-06, 4.22477e+01, 1.10692e-02],
    [-1.11757e+09, 3.97446e+08, -1.73880e+03, 4.22477e+01, 1.18004e+09, 7.63258e+05],
    [-3.86419e+06, 8.33542e+05, -6.23024e-01, 1.10692e-02, 7.63258e+05, 1.15600e+03]
])

# Invert the Fisher matrix to get the covariance matrix
cov_matrix = np.linalg.inv(fisher_matrix)

# Example parameter estimates (replace with actual MLE estimates)
parameter_estimates = np.array([1.4, 0.25, 0.5, 0.1, 0.0, 3.0])  # Replace with actual parameter estimates

# Generate samples from the posterior distribution
num_samples = 10000
samples = multivariate_normal.rvs(mean=parameter_estimates, cov=cov_matrix, size=num_samples)

# Plot the posterior distributions
param_names = ['M_c', 'eta', 'Lambda', 'delta Lambda', 't_c', 'phi_c']

fig, axes = plt.subplots(len(param_names), 1, figsize=(10, 12))
for i, param in enumerate(param_names):
    ax = axes[i]
    ax.hist(samples[:, i], bins=30, alpha=0.5, label=f'Posterior of {param}')
    ax.axvline(parameter_estimates[i], color='r', linestyle='dashed', linewidth=2, label='MLE')
    ax.set_title(f'Posterior distribution and MLE for {param}')
    ax.legend()

plt.tight_layout()
plt.show()

# Plot the heatmap of the covariance matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cov_matrix, annot=True, fmt=".2e", xticklabels=param_names, yticklabels=param_names, cmap="coolwarm")
plt.title("Heatmap of Covariance Matrix")
plt.show()
```
