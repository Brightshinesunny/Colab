```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
mass1 = 1.4 * 1.989e30  # Mass of neutron star (kg)
mass2 = 5.0 * 1.989e30  # Mass of black hole (kg)
separation = 1.0e8  # Initial separation (m)

# Initial positions
pos1 = np.array([-separation * mass2 / (mass1 + mass2), 0.0])  # Neutron star
pos2 = np.array([separation * mass1 / (mass1 + mass2), 0.0])  # Black hole

# Calculate the magnitude of the orbital velocity for circular orbit
orbital_velocity = np.sqrt(G * (mass1 + mass2) / separation)

# Initial velocities (perpendicular to the separation vector)
vel1 = np.array([0.0, orbital_velocity * mass2 / (mass1 + mass2)])  # Neutron star initial velocity (m/s)
vel2 = np.array([0.0, -orbital_velocity * mass1 / (mass1 + mass2)])  # Black hole initial velocity (m/s)

# Equations of motion for the binary system
def equations_of_motion(t, y):
    r1 = y[:2]
    v1 = y[2:4]
    r2 = y[4:6]
    v2 = y[6:8]

    r12 = r2 - r1
    distance = np.linalg.norm(r12)

    force = G * mass1 * mass2 / distance**2
    force_vector = force * r12 / distance

    dr1dt = v1
    dv1dt = force_vector / mass1
    dr2dt = v2
    dv2dt = -force_vector / mass2

    return np.concatenate((dr1dt, dv1dt, dr2dt, dv2dt))

# Initial conditions
initial_conditions = np.concatenate((pos1, vel1, pos2, vel2))

# Time span for the simulation
t_span = (0, 3e7)  # Increased time span
t_eval = np.linspace(0, 3e7, 10000)  # Increased number of evaluation points

# Solve the ODE
solution = solve_ivp(equations_of_motion, t_span, initial_conditions, t_eval=t_eval, rtol=1e-9, atol=1e-9)

# Extract the results
r1_sol = solution.y[:2]
r2_sol = solution.y[4:6]

# Plot the orbits
plt.figure(figsize=(10, 10))
plt.plot(r1_sol[0], r1_sol[1], label='Neutron Star Orbit')
plt.plot(r2_sol[0], r2_sol[1], label='Black Hole Orbit')
plt.xlabel('x position (m)')
plt.ylabel('y position (m)')
plt.legend()
plt.title('BH-NS Binary System Orbit')
plt.grid()
plt.show()
```
