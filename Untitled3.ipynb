```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate waveform data
time = np.linspace(0, 100, 1000)
inspiral = np.sin(0.1 * time**2)
merger = np.sin(0.1 * time**2) * np.exp(-0.01 * (time - 70)**2)
ringdown = np.exp(-0.1 * (time - 80)) * (time > 80)

# Combine phases
waveform = np.where(time < 70, inspiral, np.where(time < 80, merger, ringdown))

# Plot
plt.figure(figsize=(10, 6))
plt.plot(time, waveform, label="IMRPhenomPv2")
plt.axvline(x=70, color='r', linestyle='--', label="Merger")
plt.axvline(x=80, color='g', linestyle='--', label="Ringdown Start")
plt.title("IMRPhenomPv2 Waveform")
plt.xlabel("Time")
plt.ylabel("Gravitational Wave Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Simulate phase evolution data
time = np.linspace(0, 100, 1000)
phase_without_tidal = 0.1 * time**2
phase_with_tidal = 0.1 * time**2 * (1 + 0.001 * time)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(time, phase_without_tidal, label="Without Tidal Effects")
plt.plot(time, phase_with_tidal, label="With Tidal Effects", linestyle='--')
plt.title("NRTidalv2 Waveform: Phase Evolution")
plt.xlabel("Time")
plt.ylabel("Gravitational Wave Phase")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Simulate EOS data
density = np.linspace(1e14, 1e15, 100)
pressure_gpeos = density**2 * 1e-28 + np.random.normal(scale=0.1e30, size=density.size)
pressure_bsk24 = density**2 * 1e-28

# Plot
plt.figure(figsize=(10, 6))
plt.plot(density, pressure_gpeos, label="GP-EOS", linestyle='--')
plt.plot(density, pressure_bsk24, label="BSK24")
plt.fill_between(density, pressure_gpeos - 0.5e30, pressure_gpeos + 0.5e30, alpha=0.3)
plt.title("Equation of State Models")
plt.xlabel("Density (g/cm)")
plt.ylabel("Pressure (dyn/cm)")
plt.legend()
plt.grid(True)
plt.show()
```
